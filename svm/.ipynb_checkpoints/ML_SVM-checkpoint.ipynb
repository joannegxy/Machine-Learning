载入mnist数据库，将数据随机地分成训练集与测试集，训练集占原数据集的80%，而测试集则占20%，并对数据进行预处理：

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

digits = load_digits()

X,Y = digits.data, digits.target
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
scaling = MinMaxScaler(feature_range=(-1, 1)).fit(X_train)
X_train = scaling.transform(X_train)
X_test = scaling.transform(X_test)
```

首先尝试了linear svm算法：

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import seaborn as sns


linear_svm = SVC(kernel='linear')
linear_svm.fit(X_train, Y_train)
Y_pred = linear_svm.predict(X_test)
```

将linear svm的准确率和混淆矩阵输出：

```python
import matplotlib.pyplot as plt
conf_matrix=metrics.confusion_matrix(y_true=Y_test, y_pred=Y_pred)

plt.figure(figsize=(9,9))
sns.heatmap(conf_matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r',mask=conf_matrix==0);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(metrics.accuracy_score(y_true=Y_test, y_pred=Y_pred))
plt.title(all_sample_title, size = 15);
```

从上可见linear svm 准确率为0.974，判错率最高的是数字“8”判断成“1”.

接着尝试rbf kernel的svm算法:

```python
rbf_svm = SVC(kernel='rbf')
rbf_svm.fit(X_train, Y_train)
Y_pred = rbf_svm.predict(X_test)
```

将其准确率和混淆矩阵输出：

```python
conf_matrix=metrics.confusion_matrix(y_true=Y_test, y_pred=Y_pred)

plt.figure(figsize=(9,9))
sns.heatmap(conf_matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r',mask=conf_matrix==0);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(metrics.accuracy_score(y_true=Y_test, y_pred=Y_pred))
plt.title(all_sample_title, size = 15);
```

用5-fold cross validation对指定的gamma和c参数值进行训练，并计算分数：

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

folds = KFold(n_splits = 5, shuffle = True, random_state = 10)
hyperparams = [ {'gamma': [1e-2, 1e-3, 1e-4], 'C': [5,10,15,20,25,30,40,50,60,80,100,120,140,170,200,300,400,500,600,700,800]}]

rbf_svm_cv = GridSearchCV(estimator = SVC(kernel="rbf"), 
                        param_grid = hyperparams, 
                        scoring= 'accuracy', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)      
rbf_svm_cv.fit(X_train, Y_train)
```

将训练结果输出：

```python
import pandas as pd
cv_results = pd.DataFrame(rbf_svm_cv.cv_results_)
cv_results
```

将表格可视化：

```python
cv_results['param_C'] = cv_results['param_C'].astype('int')

plt.figure(figsize=(16,8))

# subplot 1/3
plt.subplot(1,3,1)
gamma_0001 = cv_results[cv_results['param_gamma']==0.0001]

plt.plot(gamma_0001["param_C"], gamma_0001["mean_test_score"])
plt.plot(gamma_0001["param_C"], gamma_0001["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=0.0001")
plt.ylim([0.60, 1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')


# subplot 2/3
plt.subplot(1,3,2)
gamma_001 = cv_results[cv_results['param_gamma']==0.001]

plt.plot(gamma_001["param_C"], gamma_001["mean_test_score"])
plt.plot(gamma_001["param_C"], gamma_001["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=0.001")
plt.ylim([0.60, 1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')

# subplot 3/3
plt.subplot(1,3,3)
gamma_01 = cv_results[cv_results['param_gamma']==0.01]

plt.plot(gamma_01["param_C"], gamma_01["mean_test_score"])
plt.plot(gamma_01["param_C"], gamma_01["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=0.01")
plt.ylim([0.60, 1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')
```

从上可见gamma取0.01的准确率比其余两个参数值的准确率大。此外，C取得太大会形成过拟合现象，训练集准确率越来越高但测试集准确率越来越低。

因此固定gamma=0.01，寻找最佳的C：

```python
C_list=np.linspace(1,50,500)
C_acc_results=[]
for c in C_list:
    best_rbf_model = SVC(C=c, gamma=0.01, kernel="rbf")
    best_rbf_model.fit(X_train, Y_train)
    Y_pred = best_rbf_model.predict(X_test)
    C_acc_results.append(metrics.accuracy_score(y_true=Y_test, y_pred=Y_pred))
C_results=dict(zip(C_list,C_acc_results))
print("optimal c:",max(C_results,key=C_results.get),"accuracy:",max(C_results.values()))
```

从上可见gamma为0.01的rbf kernel svm来说，最佳的C为6.400801603206413,其准确率为0.994.

训练gamma为0.01，c=6.400801603206413的rbf kernel svm，然后将其混淆矩阵输出：

```python
best_rbf_model = SVC(C=max(C_results,key=C_results.get), gamma=0.01, kernel="rbf")
best_rbf_model.fit(X_train, Y_train)
Y_pred = best_rbf_model.predict(X_test)
conf_matrix=metrics.confusion_matrix(y_true=Y_test, y_pred=Y_pred)
```

```python
conf_matrix=metrics.confusion_matrix(y_true=Y_test, y_pred=Y_pred)

plt.figure(figsize=(9,9))
sns.heatmap(conf_matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r',mask=conf_matrix==0);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(metrics.accuracy_score(y_true=Y_test, y_pred=Y_pred))
plt.title(all_sample_title, size = 15);
```

类似地，训练polynomial kernel svm：

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

folds = KFold(n_splits = 5, shuffle = True, random_state = 10)
hyperparams = [ {'degree': [2, 3, 4], 'C': [0.001,0.005, 0.01, 0.05, 0.1,1,10]}]

poly_svm_cv = GridSearchCV(estimator = SVC(kernel="poly"), 
                        param_grid = hyperparams, 
                        scoring= 'accuracy', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)      
poly_svm_cv.fit(X_train, Y_train)
```

```python
cv_results = pd.DataFrame(poly_svm_cv.cv_results_)
cv_results
```

```python
cv_results['param_C'] = cv_results['param_C'].astype('int')

plt.figure(figsize=(16,8))

# subplot 1/3
plt.subplot(1,3,1)
degree_2 = cv_results[cv_results['param_degree']==2]

plt.plot(degree_2["param_C"], degree_2["mean_test_score"])
plt.plot(degree_2["param_C"], degree_2["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("degree=2")
plt.ylim([0.90, 1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')


# subplot 2/3
plt.subplot(1,3,2)
degree_3 = cv_results[cv_results['param_degree']==3]

plt.plot(degree_3["param_C"], degree_3["mean_test_score"])
plt.plot(degree_3["param_C"], degree_3["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("degree=3")
plt.ylim([0.90, 1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')


# subplot 3/3
plt.subplot(1,3,3)
degree_4 = cv_results[cv_results['param_degree']==4]

plt.plot(degree_4["param_C"], degree_4["mean_test_score"])
plt.plot(degree_4["param_C"], degree_4["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("degree=4")
plt.ylim([0.90, 1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
```

固定d=3，寻找最佳的c：

```python
C_list=np.linspace(0.1,20,500)
C_acc_results=[]
for c in C_list:
    best_poly_model = SVC(C=c, degree=3, kernel="poly")
    best_poly_model.fit(X_train, Y_train)
    Y_pred = best_poly_model.predict(X_test)
    C_acc_results.append(metrics.accuracy_score(y_true=Y_test, y_pred=Y_pred))
C_results=dict(zip(C_list,C_acc_results))
print("optimal c:",max(C_results,key=C_results.get),"accuracy:",max(C_results.values()))
```

训练d=3，c=0.8178356713426852的polynomial kernel svm，然后将其混淆矩阵输出：

```python
best_poly_model = SVC(C=max(C_results,key=C_results.get), degree=3, kernel="poly")
best_poly_model.fit(X_train, Y_train)
Y_pred = best_poly_model.predict(X_test)
conf_matrix=metrics.confusion_matrix(y_true=Y_test, y_pred=Y_pred)
```
